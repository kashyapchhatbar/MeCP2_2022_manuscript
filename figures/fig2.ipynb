# Screenshots

log2 ChIP/Input and log2 ChIP-WT/ChIP-KO bigwig tracks are provided in chipseq folder. In order to visualize the tracks, open up your favourite genome browser and load the tracks and jump to the following coordinates:
- A) chr4:153558029-153676791 (mm9)
- B) chr5:95751179-110732516 (mm9)

Or you can use [pyGenomeTracks](https://pygenometracks.readthedocs.io/en/latest/index.html) to replicate the figures in the manuscript

```bash
$ cd data

$ pyGenomeTracks --tracks fig2a.ini --region chr4:153558029-153676791 -o fig2a.png --plotWidth 10 --dpi 150

$ pyGenomeTracks --tracks fig2b.ini --region chr5:95751179-110732516 -o fig2b.png --plotWidth 10 --dpi 150
```
