```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from typing import Optional
import matplotlib

sns.set_context("notebook", font_scale=1)
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['font.sans-serif'] = 'Nimbus Sans'

%matplotlib inline
%config InlineBackend.figure_formats = ['retina']
```

```python
df = pd.read_excel("data/Raw data.xlsx", usecols=[0,4,5,6], index_col=0)
df.dropna()

fig, ax = plt.subplots(figsize=(3,1.5))
sns.stripplot(data=df.dropna().T, size=5, edgecolor="white", color="black", linewidth=1, ax=ax)
sem = df.dropna().T.sem()
std = df.dropna().T.std()
mean = df.dropna().T.mean()
ax.bar(range(len(sem.index)), mean, width=0.65, color="lightgrey", edgecolor="black", linewidth=1)
ax.set_ylabel("MeCP2 / Input / mCAC")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_yticks([0, 0.5, 1, 1.5, 2])
_ = ax.text(0.0, 1.2, "Fig 3B", transform=fig.transFigure, va='top', weight="bold", fontsize="x-large")
fig.savefig("pdfs/fig3b.pdf")
```
